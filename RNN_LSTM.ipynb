```python
import os
import re
import xml.etree.ElementTree as ET
from collections import Counter
import random
from typing import List, Tuple, Dict

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
import jieba
from nltk.tokenize import word_tokenize
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 设置随机种子保证可重复性
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# ==================== 1. collate函数 ====================
def collate_fn(batch):
    """collate函数来处理变长序列"""
    sequences = [item['sequence'] for item in batch]
    labels = [item['label'] for item in batch]
    lengths = [item['length'] for item in batch]
    
    # 按长度降序排序（pack_padded_sequence需要）
    sorted_indices = sorted(range(len(lengths)), key=lambda i: lengths[i], reverse=True)
    sequences = [sequences[i] for i in sorted_indices]
    labels = [labels[i] for i in sorted_indices]
    lengths = [lengths[i] for i in sorted_indices]
    
    # 堆叠张量
    sequences_padded = torch.nn.utils.rnn.pad_sequence(sequences, batch_first=True, padding_value=0)
    labels = torch.stack(labels)
    lengths = torch.stack(lengths)
    
    return {
        'sequence': sequences_padded,
        'label': labels,
        'length': lengths
    }

# ==================== 2. 数据预处理模块 ====================
class TextProcessor:
    """文本预处理类"""
    
    def __init__(self, language='cn'):
        """
        初始化文本处理器
        Args:
            language: 'cn' 中文, 'en' 英文
        """
        self.language = language
        self.vocab = {'<PAD>': 0, '<UNK>': 1, '<SOS>': 2, '<EOS>': 3}
        self.reverse_vocab = {v: k for k, v in self.vocab.items()}
        self.max_len = 0
        self.word_freq = Counter()
        
    def tokenize(self, text: str) -> List[str]:
        """分词函数"""
        text = self.clean_text(text)
        if self.language == 'cn':
            # 中文分词
            tokens = list(jieba.cut(text))
        else:
            # 英文分词
            try:
                tokens = word_tokenize(text.lower())
            except:
                # 如果分词失败，使用简单空格分词
                tokens = text.lower().split()
        return tokens
    
    def clean_text(self, text: str) -> str:
        """清洗文本"""
        # 移除多余空白字符
        text = re.sub(r'\s+', ' ', text)
        # 移除特殊字符但保留基本标点
        if self.language == 'cn':
            text = re.sub(r'[^\u4e00-\u9fa5a-zA-Z0-9，。！？,.!?\s]', '', text)
        else:
            text = re.sub(r'[^a-zA-Z0-9,.!?\s]', '', text)
        return text.strip()
    
    def build_vocab(self, texts: List[str], min_freq=3, max_vocab_size=20000):
        """构建词汇表"""
        self.word_freq = Counter()
        
        # 统计词频
        for text in tqdm(texts, desc="构建词汇表中"):
            tokens = self.tokenize(text)
            self.word_freq.update(tokens)
            if tokens:  # 避免空序列
                self.max_len = max(self.max_len, len(tokens))
        
        # 确保最大长度至少为1
        self.max_len = max(self.max_len, 1)
        
        # 过滤低频词，构建词汇表
        sorted_words = [word for word, freq in self.word_freq.most_common() 
                       if freq >= min_freq]
        
        # 限制词汇表大小
        if max_vocab_size:
            sorted_words = sorted_words[:max_vocab_size - len(self.vocab)]
        
        # 添加单词到词汇表
        for idx, word in enumerate(sorted_words, start=len(self.vocab)):
            self.vocab[word] = idx
        
        # 更新反向词汇表
        self.reverse_vocab = {v: k for k, v in self.vocab.items()}
        
        print(f"词汇表大小: {len(self.vocab)}")
        print(f"最大序列长度: {self.max_len}")
    
    def text_to_sequence(self, text: str, max_len=None) -> torch.Tensor:
        """将文本转换为索引序列"""
        if max_len is None:
            max_len = self.max_len
        
        tokens = self.tokenize(text)
        sequence = [self.vocab.get(token, self.vocab['<UNK>']) for token in tokens]
        
        # 确保序列不为空
        if len(sequence) == 0:
            sequence = [self.vocab['<PAD>']]
        
        # 截断
        if len(sequence) > max_len:
            sequence = sequence[:max_len]
        
        return torch.tensor(sequence, dtype=torch.long)
    
    def sequence_to_text(self, sequence: List[int]) -> str:
        """将索引序列转换回文本"""
        return ' '.join([self.reverse_vocab.get(idx, '<UNK>') for idx in sequence 
                        if idx not in {0, 1, 2, 3}])  # 忽略特殊标记

# ==================== 3. 数据加载模块 ====================
def parse_reviews(file_path: str) -> List[Tuple[str, int]]:
    """
    解析训练集文件
    返回: [(文本, 标签), ...]
    """
    reviews = []
    
    # 尝试多种编码方式读取文件
    encodings_to_try = ['utf-8-sig', 'utf-8', 'gbk', 'gb2312', 'latin-1']
    
    for encoding in encodings_to_try:
        try:
            with open(file_path, 'r', encoding=encoding) as f:
                content = f.read()
            print(f"使用编码 {encoding} 成功读取文件: {file_path}")
            break
        except UnicodeDecodeError:
            continue
    else:
        # 如果所有编码都失败，尝试二进制读取
        with open(file_path, 'rb') as f:
            content_bytes = f.read()
        # 尝试解码，忽略错误
        try:
            content = content_bytes.decode('utf-8', errors='ignore')
        except:
            content = content_bytes.decode('latin-1', errors='ignore')
        print(f"使用二进制读取并忽略错误方式读取文件: {file_path}")
    
    # 使用正则表达式提取review
    pattern = r'<review id="(\d+)">\s*(.*?)\s*</review>'
    matches = re.findall(pattern, content, re.DOTALL)
    
    for review_id, text in matches:
        # 根据文件名判断标签
        if 'negative' in file_path:
            label = 0
        elif 'positive' in file_path:
            label = 1
        else:
            raise ValueError(f"无法从文件名判断标签: {file_path}")
        
        # 清理文本
        text = text.strip()
        if text:  # 只添加非空文本
            reviews.append((text, label))
    
    return reviews

def parse_test_data(text_file: str, label_file: str) -> List[Tuple[str, int]]:
    """
    解析测试集文件和标签文件
    """
    reviews = []
    
    def read_file_with_fallback(filepath):
        """尝试多种编码方式读取文件"""
        encodings_to_try = ['utf-8-sig', 'utf-8', 'gbk', 'gb2312', 'latin-1']
        
        for encoding in encodings_to_try:
            try:
                with open(filepath, 'r', encoding=encoding) as f:
                    content = f.read()
                print(f"使用编码 {encoding} 成功读取文件: {filepath}")
                return content
            except UnicodeDecodeError:
                continue
        
        # 如果所有编码都失败，尝试二进制读取
        with open(filepath, 'rb') as f:
            content_bytes = f.read()
        
        # 尝试解码，忽略错误
        for encoding in ['utf-8', 'gbk', 'latin-1']:
            try:
                content = content_bytes.decode(encoding, errors='ignore')
                print(f"使用二进制解码 {encoding} 并忽略错误方式读取文件: {filepath}")
                return content
            except:
                continue
        
        # 如果所有解码都失败，使用latin-1（不会失败）
        content = content_bytes.decode('latin-1', errors='ignore')
        print(f"使用latin-1解码并忽略错误方式读取文件: {filepath}")
        return content
    
    # 读取文本文件
    text_content = read_file_with_fallback(text_file)
    
    # 读取标签文件
    label_content = read_file_with_fallback(label_file)
    
    # 提取文本
    text_pattern = r'<review id="(\d+)">\s*(.*?)\s*</review>'
    text_matches = re.findall(text_pattern, text_content, re.DOTALL)
    
    # 提取标签
    label_pattern = r'<review id="(\d+)"\s+label="(\d+)"'
    label_matches = re.findall(label_pattern, label_content, re.DOTALL)
    
    # 创建文本和标签的映射
    text_dict = {}
    for id_, text in text_matches:
        try:
            text_dict[int(id_)] = text.strip()
        except (ValueError, KeyError):
            continue
    
    label_dict = {}
    for id_, label in label_matches:
        try:
            label_dict[int(id_)] = int(label)
        except (ValueError, KeyError):
            continue
    
    # 确保顺序一致
    for review_id in sorted(text_dict.keys()):
        if review_id in label_dict and text_dict[review_id]:
            reviews.append((text_dict[review_id], label_dict[review_id]))
    
    print(f"成功解析 {len(reviews)} 条测试数据")
    return reviews

# ==================== 4. 数据集类 ====================
class SentimentDataset(Dataset):
    """情感分析数据集类"""
    
    def __init__(self, texts: List[str], labels: List[int], processor: TextProcessor, max_len=100):
        self.texts = texts
        self.labels = labels
        self.processor = processor
        self.max_len = max_len
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        
        # 转换为序列
        sequence = self.processor.text_to_sequence(text, self.max_len)
        tokens = self.processor.tokenize(text)
        length = min(len(tokens), self.max_len)
        
        # 确保长度至少为1
        if length == 0:
            length = 1
        
        return {
            'sequence': sequence,
            'label': torch.tensor(label, dtype=torch.long),
            'length': torch.tensor(length, dtype=torch.long)
        }

# ==================== 5. LSTM模型 ====================
class SentimentLSTM(nn.Module):
    """情感分析LSTM模型"""
    
    def __init__(self, vocab_size, embedding_dim=200, hidden_dim=256, 
                 output_dim=2, n_layers=2, dropout=0.5, padding_idx=0):
        super().__init__()
        
        # 嵌入层
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=padding_idx)
        
        # LSTM层
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers,
                           bidirectional=True, dropout=dropout, batch_first=True)
        
        # 全连接层
        self.fc = nn.Linear(hidden_dim * 2, output_dim)  # *2 因为双向
        
        # Dropout
        self.dropout = nn.Dropout(dropout)
        
        # 初始化权重
        self._init_weights()
    
    def _init_weights(self):
        """初始化权重"""
        nn.init.normal_(self.embedding.weight, mean=0, std=0.1)
        for name, param in self.lstm.named_parameters():
            if 'weight' in name:
                nn.init.orthogonal_(param)
            elif 'bias' in name:
                nn.init.constant_(param, 0.0)
        nn.init.xavier_normal_(self.fc.weight)
        nn.init.constant_(self.fc.bias, 0.0)
    
    def forward(self, text, text_lengths):
        # text: [batch_size, seq_len]
        # text_lengths: [batch_size]
        
        # 嵌入
        embedded = self.dropout(self.embedding(text))
        # embedded: [batch_size, seq_len, embedding_dim]
        
        # 打包序列以处理可变长度
        packed_embedded = nn.utils.rnn.pack_padded_sequence(
            embedded, text_lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        
        # LSTM
        packed_output, (hidden, cell) = self.lstm(packed_embedded)
        
        # 解包
        output, output_lengths = nn.utils.rnn.pad_packed_sequence(packed_output, batch_first=True)
        
        # 获取最后一个时间步的输出
        # 双向LSTM，将前向和后向的最后一个隐藏状态连接起来
        hidden = self.dropout(torch.cat((hidden[-2, :, :], hidden[-1, :, :]), dim=1))
        
        # 全连接层
        return self.fc(hidden)

# ==================== 6. 训练和评估模块 ====================
class SentimentTrainer:
    """情感分析训练器"""
    
    def __init__(self, model, device, model_save_path):
        self.model = model
        self.device = device
        self.model_save_path = model_save_path
        self.train_losses = []
        self.val_accuracies = []
        self.train_accuracies = []
        
    def train_epoch(self, train_loader, criterion, optimizer, epoch, n_epochs):
        """训练一个epoch"""
        self.model.train()
        epoch_loss = 0
        correct = 0
        total = 0
        
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{n_epochs} [训练]")
        for batch in pbar:
            sequences = batch['sequence'].to(self.device)
            labels = batch['label'].to(self.device)
            lengths = batch['length'].to(self.device)
            
            # 确保长度不为0
            lengths = torch.clamp(lengths, min=1)
            
            optimizer.zero_grad()
            
            # 前向传播
            predictions = self.model(sequences, lengths)
            
            # 计算损失
            loss = criterion(predictions, labels)
            
            # 反向传播
            loss.backward()
            
            # 梯度裁剪
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            
            optimizer.step()
            
            # 统计
            epoch_loss += loss.item()
            _, predicted = torch.max(predictions.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            # 更新进度条
            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'acc': f'{100 * correct/total:.2f}%'
            })
        
        avg_loss = epoch_loss / len(train_loader)
        accuracy = 100 * correct / total
        
        self.train_losses.append(avg_loss)
        self.train_accuracies.append(accuracy)
        
        return avg_loss, accuracy
    
    def evaluate(self, val_loader, criterion):
        """评估模型"""
        self.model.eval()
        val_loss = 0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for batch in tqdm(val_loader, desc="评估"):
                sequences = batch['sequence'].to(self.device)
                labels = batch['label'].to(self.device)
                lengths = batch['length'].to(self.device)
                
                # 确保长度不为0
                lengths = torch.clamp(lengths, min=1)
                
                predictions = self.model(sequences, lengths)
                loss = criterion(predictions, labels)
                
                val_loss += loss.item()
                _, predicted = torch.max(predictions.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        
        avg_loss = val_loss / len(val_loader)
        accuracy = 100 * correct / total
        
        return avg_loss, accuracy
    
    def save_model(self, accuracy, epoch, processor, config, language):
        """保存模型"""
        if not os.path.exists(self.model_save_path):
            os.makedirs(self.model_save_path)
        
        model_path = os.path.join(self.model_save_path, f'{language}_model_epoch{epoch}_acc{accuracy:.2f}.pth')
        torch.save({
            'epoch': epoch,
            'model_state_dict': self.model.state_dict(),
            'processor_vocab': processor.vocab,
            'processor_max_len': processor.max_len,
            'config': config,
            'language': language,
            'accuracy': accuracy,
        }, model_path)
        print(f"模型已保存到: {model_path}")
        return model_path
    
    def plot_training_history(self, language):
        """绘制训练历史"""
        if len(self.train_losses) == 0 or len(self.val_accuracies) == 0:
            print(f"没有足够的数据绘制{language}训练历史")
            return
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # 损失曲线
        ax1.plot(self.train_losses, label='训练损失', color='blue')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('损失')
        ax1.set_title(f'{language}模型训练损失曲线')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 准确率曲线
        epochs = range(1, len(self.train_accuracies) + 1)
        ax2.plot(epochs, self.train_accuracies, label='训练准确率', color='green', marker='o', alpha=0.7)
        ax2.plot(epochs, self.val_accuracies, label='验证准确率', color='red', marker='s', alpha=0.7)
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('准确率 (%)')
        ax2.set_title(f'{language}模型准确率曲线')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'{language}_training_history.png', dpi=300, bbox_inches='tight')
        plt.show()

# ==================== 7. 主训练函数 ====================
def train_language_model(language, data_paths, config):
    """
    训练指定语言的模型
    Args:
        language: 'cn' 或 'en'
        data_paths: 数据路径字典
        config: 训练配置
    """
    print(f"\n{'='*60}")
    print(f"开始训练{language}模型")
    print(f"{'='*60}\n")
    
    # 1. 加载数据
    print("1. 加载数据...")
    
    # 加载训练数据
    pos_data = parse_reviews(data_paths['train_pos'])
    neg_data = parse_reviews(data_paths['train_neg'])
    
    # 合并训练数据
    all_train_data = pos_data + neg_data
    random.shuffle(all_train_data)
    
    # 分离文本和标签
    train_texts = [item[0] for item in all_train_data]
    train_labels = [item[1] for item in all_train_data]
    
    print(f"训练集大小: {len(train_texts)}")
    print(f"正面样本: {train_labels.count(1)}, 负面样本: {train_labels.count(0)}")
    
    # 加载测试数据
    test_data = parse_test_data(data_paths['test_text'], data_paths['test_label'])
    test_texts = [item[0] for item in test_data]
    test_labels = [item[1] for item in test_data]
    
    print(f"测试集大小: {len(test_texts)}")
    print(f"正面样本: {test_labels.count(1)}, 负面样本: {test_labels.count(0)}")
    
    # 2. 创建文本处理器并构建词汇表
    print("\n2. 构建词汇表...")
    processor = TextProcessor(language=language)
    processor.build_vocab(train_texts, 
                         min_freq=config['min_freq'], 
                         max_vocab_size=config['max_vocab_size'])
    
    # 3. 创建数据集和数据加载器
    print("\n3. 创建数据加载器...")
    
    # 分割训练集和验证集
    split_idx = int(len(train_texts) * config['train_ratio'])
    
    train_dataset = SentimentDataset(
        train_texts[:split_idx], 
        train_labels[:split_idx], 
        processor, 
        max_len=config['max_len']
    )
    
    val_dataset = SentimentDataset(
        train_texts[split_idx:], 
        train_labels[split_idx:], 
        processor, 
        max_len=config['max_len']
    )
    
    test_dataset = SentimentDataset(
        test_texts, 
        test_labels, 
        processor, 
        max_len=config['max_len']
    )
    
    train_loader = DataLoader(
        train_dataset, 
        batch_size=config['batch_size'], 
        shuffle=True,
        collate_fn=collate_fn
    )
    
    val_loader = DataLoader(
        val_dataset, 
        batch_size=config['batch_size'], 
        shuffle=False,
        collate_fn=collate_fn
    )
    
    test_loader = DataLoader(
        test_dataset, 
        batch_size=config['batch_size'], 
        shuffle=False,
        collate_fn=collate_fn
    )
    
    # 4. 创建模型
    print("\n4. 创建模型...")
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"使用设备: {device}")
    
    vocab_size = len(processor.vocab)
    model = SentimentLSTM(
        vocab_size=vocab_size,
        embedding_dim=config['embedding_dim'],
        hidden_dim=config['hidden_dim'],
        output_dim=config['output_dim'],
        n_layers=config['n_layers'],
        dropout=config['dropout']
    ).to(device)
    
    # 打印模型参数数量
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"总参数数量: {total_params:,}")
    print(f"可训练参数数量: {trainable_params:,}")
    
    # 5. 定义损失函数和优化器
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=config['learning_rate'])
    
    # 使用ReduceLROnPlateau学习率调度器
    # 注意：有些PyTorch版本可能需要指定verbose参数，有些则不需要
    # 这里使用try-except来处理不同版本
    try:
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='max', factor=0.5, patience=2, verbose=True
        )
    except TypeError:
        # 如果当前版本不支持verbose参数
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='max', factor=0.5, patience=2
        )
        print("注意: 当前PyTorch版本的ReduceLROnPlateau不支持verbose参数")
    
    # 6. 创建训练器
    trainer = SentimentTrainer(
        model=model,
        device=device,
        model_save_path=config['model_save_path']
    )
    
    # 7. 训练模型
    print(f"\n5. 开始训练{language}模型...")
    print(f"训练计划:")
    print(f"  - 总轮次: {config['n_epochs']}")
    print(f"  - 批次大小: {config['batch_size']}")
    print(f"  - 学习率: {config['learning_rate']}")
    print(f"  - 训练集大小: {len(train_dataset)}")
    print(f"  - 验证集大小: {len(val_dataset)}")
    print(f"  - 测试集大小: {len(test_dataset)}")
    print(f"  - 词汇表大小: {vocab_size}")
    print(f"  - 序列最大长度: {config['max_len']}")
    
    best_accuracy = 0.0
    best_model_path = None
    patience_counter = 0
    
    for epoch in range(config['n_epochs']):
        print(f"\n{'='*50}")
        print(f"轮次 {epoch+1}/{config['n_epochs']}")
        print(f"{'='*50}")
        
        # 训练
        train_loss, train_acc = trainer.train_epoch(
            train_loader, criterion, optimizer, epoch, config['n_epochs']
        )
        print(f"训练损失: {train_loss:.4f}, 训练准确率: {train_acc:.2f}%")
        
        # 验证
        val_loss, val_accuracy = trainer.evaluate(val_loader, criterion)
        trainer.val_accuracies.append(val_accuracy)
        print(f"验证损失: {val_loss:.4f}, 验证准确率: {val_accuracy:.2f}%")
        
        # 学习率调整
        scheduler.step(val_accuracy)
        
        # 打印当前学习率
        current_lr = optimizer.param_groups[0]['lr']
        print(f"当前学习率: {current_lr}")
        
        # 保存最佳模型
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            best_model_path = trainer.save_model(val_accuracy, epoch, processor, config, language)
            patience_counter = 0
            print(f"✓ 新的最佳准确率: {val_accuracy:.2f}%")
        else:
            patience_counter += 1
            print(f"准确率未提升，耐心计数: {patience_counter}/{config['patience']}")
        
        # 早停
        if patience_counter >= config['patience']:
            print(f"早停触发，停止训练")
            break
    
    # 8. 在测试集上评估最佳模型
    print(f"\n6. 在测试集上评估最佳模型...")
    if best_model_path and os.path.exists(best_model_path):
        # 加载最佳模型
        checkpoint = torch.load(best_model_path, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        print(f"已加载最佳模型: {best_model_path}")
    
    test_loss, test_accuracy = trainer.evaluate(test_loader, criterion)
    print(f"测试集准确率: {test_accuracy:.2f}%")
    
    # 9. 绘制训练历史
    print(f"\n7. 绘制训练历史...")
    trainer.plot_training_history("中文" if language == 'cn' else "英文")
    
    # 10. 返回处理器和最佳准确率
    return processor, test_accuracy, best_model_path

# ==================== 8. 主程序 ====================
def main():
    """主函数"""
    print("="*60)
    print("中文/英文情感分析LSTM模型训练程序")
    print("="*60)
    
    # 路径配置
    BASE_PATH = r"C:\PDwork"
    
    data_paths = {
        'cn': {
            'train_pos': os.path.join(BASE_PATH, 'cn_sample_data', 'sample.positive.txt'),
            'train_neg': os.path.join(BASE_PATH, 'cn_sample_data', 'sample.negative.txt'),
            'test_text': os.path.join(BASE_PATH, 'test.cn.txt'),
            'test_label': os.path.join(BASE_PATH, 'test.label.cn.txt')
        },
        'en': {
            'train_pos': os.path.join(BASE_PATH, 'en_sample_data', 'sample.positive.txt'),
            'train_neg': os.path.join(BASE_PATH, 'en_sample_data', 'sample.negative.txt'),
            'test_text': os.path.join(BASE_PATH, 'test.en.txt'),
            'test_label': os.path.join(BASE_PATH, 'test.label.en.txt')
        }
    }
    
    # 检查路径是否存在
    for lang, paths in data_paths.items():
        for name, path in paths.items():
            if not os.path.exists(path):
                print(f"错误: 路径不存在 - {path}")
                return
    
    # 训练配置
    config = {
        # 模型参数
        'embedding_dim': 128,  # 减小嵌入维度以降低内存使用
        'hidden_dim': 128,     # 减小隐藏层维度
        'output_dim': 2,       # 二分类
        'n_layers': 2,
        'dropout': 0.3,        # 减小dropout率
        
        # 训练参数
        'batch_size': 32,      # 批次大小
        'learning_rate': 0.001,
        'n_epochs': 15,        # 减少训练轮次
        'patience': 4,         # 早停耐心
        
        # 数据参数
        'max_len': 100,        # 减小最大序列长度
        'min_freq': 2,         # 降低最小词频
        'max_vocab_size': 15000,  # 减小词汇表大小
        'train_ratio': 0.8,    # 训练集比例
        
        # 保存路径
        'model_save_path': './saved_models'
    }
    
    # 打印训练计划
    print("\n训练计划:")
    print("-" * 40)
    print(f"1. 数据统计:")
    print(f"   - 中文训练集: 10000条 (正负各5000)")
    print(f"   - 英文训练集: 10000条 (正负各5000)")
    print(f"   - 中文测试集: 2500条")
    print(f"   - 英文测试集: 2500条")
    print(f"\n2. 模型架构:")
    print(f"   - 嵌入维度: {config['embedding_dim']}")
    print(f"   - 隐藏层维度: {config['hidden_dim']}")
    print(f"   - LSTM层数: {config['n_layers']}")
    print(f"   - Dropout率: {config['dropout']}")
    print(f"\n3. 训练参数:")
    print(f"   - 批次大小: {config['batch_size']}")
    print(f"   - 学习率: {config['learning_rate']}")
    print(f"   - 最大轮次: {config['n_epochs']}")
    print(f"   - 早停耐心: {config['patience']}")
    print(f"\n4. 数据处理:")
    print(f"   - 最大序列长度: {config['max_len']}")
    print(f"   - 最小词频: {config['min_freq']}")
    print(f"   - 最大词汇表大小: {config['max_vocab_size']}")
    print(f"   - 训练验证比例: {config['train_ratio']}:{1-config['train_ratio']:.1f}")
    print("-" * 40)
    
    # 训练中文模型
    print("\n开始训练中文模型...")
    cn_processor, cn_accuracy, cn_model_path = train_language_model('cn', data_paths['cn'], config)
    
    # 训练英文模型
    print("\n开始训练英文模型...")
    en_processor, en_accuracy, en_model_path = train_language_model('en', data_paths['en'], config)
    
    # 打印最终结果
    print("\n" + "="*60)
    print("训练完成！")
    print("="*60)
    print(f"中文模型测试准确率: {cn_accuracy:.2f}%")
    print(f"英文模型测试准确率: {en_accuracy:.2f}%")
    if cn_model_path:
        print(f"\n中文模型保存在: {cn_model_path}")
    if en_model_path:
        print(f"英文模型保存在: {en_model_path}")
    print("\n训练历史图已保存为: 中文_training_history.png 和 英文_training_history.png")
    
    # 保存词汇表
    import pickle
    with open('cn_vocab.pkl', 'wb') as f:
        pickle.dump(cn_processor.vocab, f)
    with open('en_vocab.pkl', 'wb') as f:
        pickle.dump(en_processor.vocab, f)
    print("\n词汇表已保存为: cn_vocab.pkl 和 en_vocab.pkl")

# ==================== 9. 预测函数 ====================
def predict_sentiment(text, processor, model_path, device='cpu'):
    """使用训练好的模型预测情感"""
    # 加载模型
    checkpoint = torch.load(model_path, map_location=device)
    
    # 创建模型实例
    model = SentimentLSTM(
        vocab_size=len(processor.vocab),
        embedding_dim=checkpoint['config']['embedding_dim'],
        hidden_dim=checkpoint['config']['hidden_dim'],
        output_dim=checkpoint['config']['output_dim'],
        n_layers=checkpoint['config']['n_layers'],
        dropout=checkpoint['config']['dropout']
    ).to(device)
    
    # 加载权重
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    # 处理文本
    sequence = processor.text_to_sequence(text)
    sequence = sequence.unsqueeze(0)  # 添加batch维度
    length = torch.tensor([min(len(processor.tokenize(text)), processor.max_len)], 
                         dtype=torch.long).to(device)
    
    # 确保长度至少为1
    if length.item() == 0:
        length = torch.tensor([1], dtype=torch.long).to(device)
    
    # 预测
    with torch.no_grad():
        output = model(sequence, length)
        probabilities = F.softmax(output, dim=1)
        _, predicted = torch.max(output.data, 1)
    
    sentiment = "正面" if predicted.item() == 1 else "负面"
    confidence = probabilities[0][predicted.item()].item() * 100
    
    return sentiment, confidence

# ==================== 10. 示例使用 ====================
def example_predictions():
    """示例预测函数"""
    # 中文文本
    cn_examples = [
        "这个电影真的太棒了，演员演技在线，剧情也很吸引人！",
        "产品质量很差，用了一次就坏了，非常失望。",
        "服务态度很好，送货速度快，包装也很仔细。",
        "一点都不好用，跟描述完全不符，浪费钱。"
    ]
    
    # 英文文本
    en_examples = [
        "This product is amazing! It works perfectly and I love it.",
        "Terrible experience. The product broke after just one week of use.",
        "Excellent service, fast delivery, and great quality. Highly recommended!",
        "Worst purchase ever. Complete waste of money."
    ]
    
    print("\n" + "="*60)
    print("示例预测:")
    print("="*60)
    
    print("\n注意: 需要先训练模型才能进行预测")
    print("训练完成后，可以调用 predict_sentiment() 函数进行预测")

# ==================== 11. 运行主程序 ====================
if __name__ == "__main__":
    # 检查必要的库
    try:
        import nltk
        nltk.download('punkt', quiet=True)
    except:
        print("警告: 无法下载nltk数据，英文分词可能需要手动下载")
        print("请运行: nltk.download('punkt')")
    
    # 运行主程序
    main()
    
    # 显示示例
    example_predictions()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import json
import os
from typing import Dict, List, Tuple, Any
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

def convert_to_serializable(obj: Any) -> Any:
    """将numpy数据类型转换为Python可序列化类型"""
    if isinstance(obj, np.integer):
        return int(obj)
    elif isinstance(obj, np.floating):
        return float(obj)
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, np.bool_):
        return bool(obj)
    elif isinstance(obj, dict):
        return {key: convert_to_serializable(value) for key, value in obj.items()}
    elif isinstance(obj, list):
        return [convert_to_serializable(item) for item in obj]
    elif isinstance(obj, tuple):
        return tuple(convert_to_serializable(item) for item in obj)
    else:
        return obj

class ModelPerformanceEvaluator:
    """模型性能评估器"""
    
    def __init__(self, test_loss=0.5701, confusion_matrix_data=None, language='cn'):
        """
        初始化评估器
        
        Args:
            test_loss: 测试损失
            confusion_matrix_data: 混淆矩阵数据 [[TN, FP], [FN, TP]]
            language: 语言类型 ('cn' 或 'en')
        """
        self.test_loss = test_loss
        
        # 使用提供的混淆矩阵
        if confusion_matrix_data is None:
            # 使用提供的混淆矩阵
            self.cm = np.array([[905, 345], [274, 976]])
        else:
            self.cm = np.array(confusion_matrix_data)
        
        self.language = language
        self.language_name = "中文" if language == 'cn' else "英文"
        
        # 从混淆矩阵推导预测结果
        self.y_true, self.y_pred = self._derive_predictions_from_cm()
        
        # 计算所有指标
        self.metrics = self._calculate_all_metrics()
        
    def _derive_predictions_from_cm(self) -> Tuple[np.ndarray, np.ndarray]:
        """从混淆矩阵推导真实标签和预测标签"""
        # 从混淆矩阵结构推断
        # [[TN, FP], 
        #  [FN, TP]]
        tn, fp, fn, tp = self.cm.ravel()
        
        # 创建真实标签 (0: Negative, 1: Positive)
        y_true = np.array([0] * (tn + fp) + [1] * (fn + tp))
        
        # 创建预测标签
        y_pred = np.array([0] * tn + [1] * fp + [0] * fn + [1] * tp)
        
        return y_true, y_pred
    
    def _calculate_all_metrics(self) -> Dict[str, float]:
        """计算所有评估指标"""
        metrics = {}
        
        # 基础指标
        metrics['test_loss'] = float(self.test_loss)
        metrics['accuracy'] = float(accuracy_score(self.y_true, self.y_pred))
        metrics['precision_macro'] = float(precision_score(self.y_true, self.y_pred, average='macro'))
        metrics['recall_macro'] = float(recall_score(self.y_true, self.y_pred, average='macro'))
        metrics['f1_macro'] = float(f1_score(self.y_true, self.y_pred, average='macro'))
        
        # 各类别指标
        metrics['precision_negative'] = float(precision_score(self.y_true, self.y_pred, pos_label=0, zero_division=0))
        metrics['recall_negative'] = float(recall_score(self.y_true, self.y_pred, pos_label=0, zero_division=0))
        metrics['f1_negative'] = float(f1_score(self.y_true, self.y_pred, pos_label=0, zero_division=0))
        
        metrics['precision_positive'] = float(precision_score(self.y_true, self.y_pred, pos_label=1, zero_division=0))
        metrics['recall_positive'] = float(recall_score(self.y_true, self.y_pred, pos_label=1, zero_division=0))
        metrics['f1_positive'] = float(f1_score(self.y_true, self.y_pred, pos_label=1, zero_division=0))
        
        # 计算混淆矩阵指标
        tn, fp, fn, tp = self.cm.ravel()
        metrics['true_negative'] = int(tn)
        metrics['false_positive'] = int(fp)
        metrics['false_negative'] = int(fn)
        metrics['true_positive'] = int(tp)
        
        # 计算误分类率
        metrics['error_rate'] = float(1 - metrics['accuracy'])
        metrics['false_positive_rate'] = float(fp / (fp + tn) if (fp + tn) > 0 else 0)
        metrics['false_negative_rate'] = float(fn / (fn + tp) if (fn + tp) > 0 else 0)
        
        return metrics
    
    def generate_classification_report(self) -> str:
        """生成分类报告"""
        return classification_report(
            self.y_true, 
            self.y_pred, 
            target_names=['Negative', 'Positive'],
            digits=4
        )
    
    def plot_confusion_matrix(self, save_path: str = None):
        """绘制混淆矩阵热力图"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # 左图：百分比热力图
        cm_percentage = self.cm.astype('float') / self.cm.sum(axis=1)[:, np.newaxis] * 100
        
        im1 = ax1.imshow(cm_percentage, cmap='Blues', vmin=0, vmax=100)
        ax1.set_xticks([0, 1])
        ax1.set_yticks([0, 1])
        ax1.set_xticklabels(['预测: Negative', '预测: Positive'])
        ax1.set_yticklabels(['真实: Negative', '真实: Positive'])
        
        # 添加百分比标签
        for i in range(2):
            for j in range(2):
                text = f'{cm_percentage[i, j]:.1f}%\n({self.cm[i, j]})'
                color = 'white' if cm_percentage[i, j] > 50 else 'black'
                ax1.text(j, i, text, ha='center', va='center', 
                        color=color, fontsize=12, fontweight='bold')
        
        ax1.set_title(f'{self.language_name}情感分析混淆矩阵 (百分比)', fontsize=14, fontweight='bold')
        plt.colorbar(im1, ax=ax1, label='百分比 (%)')
        
        # 右图：原始数值
        im2 = ax2.imshow(self.cm, cmap='Reds', vmin=0, vmax=max(self.cm.ravel()))
        ax2.set_xticks([0, 1])
        ax2.set_yticks([0, 1])
        ax2.set_xticklabels(['预测: Negative', '预测: Positive'])
        ax2.set_yticklabels(['真实: Negative', '真实: Positive'])
        
        # 添加数值标签
        for i in range(2):
            for j in range(2):
                ax2.text(j, i, str(self.cm[i, j]), ha='center', va='center', 
                        color='white' if self.cm[i, j] > (max(self.cm.ravel()) / 2) else 'black',
                        fontsize=14, fontweight='bold')
        
        ax2.set_title(f'{self.language_name}情感分析混淆矩阵 (原始数值)', fontsize=14, fontweight='bold')
        plt.colorbar(im2, ax=ax2, label='样本数量')
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
    
    def plot_performance_metrics(self, save_path: str = None):
        """绘制性能指标对比图"""
        # 准备指标数据
        metrics_data = {
            '指标': ['准确率', '测试损失', '精确率(Negative)', '召回率(Negative)', 'F1分数(Negative)',
                   '精确率(Positive)', '召回率(Positive)', 'F1分数(Positive)', '宏平均精确率', '宏平均召回率', '宏平均F1分数'],
            '值': [
                self.metrics['accuracy'],
                self.metrics['test_loss'],
                self.metrics['precision_negative'],
                self.metrics['recall_negative'],
                self.metrics['f1_negative'],
                self.metrics['precision_positive'],
                self.metrics['recall_positive'],
                self.metrics['f1_positive'],
                self.metrics['precision_macro'],
                self.metrics['recall_macro'],
                self.metrics['f1_macro']
            ]
        }
        
        # 创建DataFrame
        df = pd.DataFrame(metrics_data)
        
        # 绘制横向柱状图
        fig, axes = plt.subplots(1, 2, figsize=(16, 8))
        
        # 左图：主要性能指标
        main_metrics = ['准确率', '精确率(Negative)', '召回率(Negative)', 'F1分数(Negative)',
                       '精确率(Positive)', '召回率(Positive)', 'F1分数(Positive)']
        main_values = [self.metrics[m] for m in ['accuracy', 'precision_negative', 'recall_negative', 
                                                'f1_negative', 'precision_positive', 'recall_positive', 'f1_positive']]
        
        colors = ['skyblue', 'lightcoral', 'lightcoral', 'lightcoral', 
                 'lightgreen', 'lightgreen', 'lightgreen']
        
        y_pos = np.arange(len(main_metrics))
        bars = axes[0].barh(y_pos, main_values, color=colors, edgecolor='black')
        
        axes[0].set_yticks(y_pos)
        axes[0].set_yticklabels(main_metrics, fontsize=11)
        axes[0].set_xlabel('分数', fontsize=12)
        axes[0].set_title(f'{self.language_name}模型主要性能指标', fontsize=14, fontweight='bold')
        axes[0].set_xlim([0, 1.0])
        axes[0].grid(True, alpha=0.3, axis='x')
        
        # 添加数值标签
        for bar, value in zip(bars, main_values):
            width = bar.get_width()
            axes[0].text(width + 0.01, bar.get_y() + bar.get_height()/2, 
                       f'{value:.4f}', ha='left', va='center', fontweight='bold')
        
        # 右图：分类对比
        categories = ['Negative', 'Positive']
        precision = [self.metrics['precision_negative'], self.metrics['precision_positive']]
        recall = [self.metrics['recall_negative'], self.metrics['recall_positive']]
        f1 = [self.metrics['f1_negative'], self.metrics['f1_positive']]
        
        x = np.arange(len(categories))
        width = 0.25
        
        bars1 = axes[1].bar(x - width, precision, width, label='精确率', color='skyblue', edgecolor='black')
        bars2 = axes[1].bar(x, recall, width, label='召回率', color='lightgreen', edgecolor='black')
        bars3 = axes[1].bar(x + width, f1, width, label='F1分数', color='salmon', edgecolor='black')
        
        axes[1].set_xlabel('类别', fontsize=12)
        axes[1].set_ylabel('分数', fontsize=12)
        axes[1].set_title(f'{self.language_name}各类别性能对比', fontsize=14, fontweight='bold')
        axes[1].set_xticks(x)
        axes[1].set_xticklabels(categories)
        axes[1].legend()
        axes[1].set_ylim([0, 1.0])
        axes[1].grid(True, alpha=0.3, axis='y')
        
        # 添加数值标签
        for bars in [bars1, bars2, bars3]:
            for bar in bars:
                height = bar.get_height()
                axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                           f'{height:.4f}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
    
    def plot_error_analysis(self, save_path: str = None):
        """绘制误分类分析图"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
        
        # 左图：误分类饼图
        labels = ['正确分类', '错误分类']
        sizes = [
            self.metrics['true_negative'] + self.metrics['true_positive'],
            self.metrics['false_positive'] + self.metrics['false_negative']
        ]
        colors = ['lightgreen', 'lightcoral']
        explode = (0, 0.1)
        
        wedges, texts, autotexts = ax1.pie(sizes, explode=explode, labels=labels, colors=colors,
                                          autopct='%1.1f%%', startangle=90, shadow=True)
        
        for autotext in autotexts:
            autotext.set_color('black')
            autotext.set_fontweight('bold')
        
        ax1.set_title(f'{self.language_name}模型分类结果分布', fontsize=14, fontweight='bold')
        
        # 右图：详细误分类分析
        error_types = ['将Negative误判为Positive', '将Positive误判为Negative']
        error_counts = [self.metrics['false_positive'], self.metrics['false_negative']]
        error_percentages = [self.metrics['false_positive_rate'] * 100, 
                            self.metrics['false_negative_rate'] * 100]
        
        x = np.arange(len(error_types))
        ax2.bar(x, error_counts, color=['orange', 'red'], edgecolor='black', alpha=0.7)
        
        ax2.set_xlabel('误分类类型', fontsize=12)
        ax2.set_ylabel('样本数量', fontsize=12, color='black')
        ax2.set_xticks(x)
        ax2.set_xticklabels(error_types, rotation=15, ha='right')
        ax2.set_title(f'{self.language_name}模型误分类分析', fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3, axis='y')
        
        # 添加数量标签
        for i, (count, percentage) in enumerate(zip(error_counts, error_percentages)):
            ax2.text(i, count + 5, f'{count}\n({percentage:.1f}%)', 
                    ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
    
    def plot_class_distribution(self, save_path: str = None):
        """绘制类别分布对比图"""
        # 真实分布
        true_negative = self.metrics['true_negative'] + self.metrics['false_negative']
        true_positive = self.metrics['false_positive'] + self.metrics['true_positive']
        
        # 预测分布
        pred_negative = self.metrics['true_negative'] + self.metrics['false_positive']
        pred_positive = self.metrics['false_negative'] + self.metrics['true_positive']
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
        
        # 左图：真实分布
        true_labels = ['Negative', 'Positive']
        true_counts = [true_negative, true_positive]
        true_colors = ['lightblue', 'lightgreen']
        
        bars1 = ax1.bar(true_labels, true_counts, color=true_colors, edgecolor='black', alpha=0.8)
        ax1.set_xlabel('真实类别', fontsize=12)
        ax1.set_ylabel('样本数量', fontsize=12)
        ax1.set_title(f'{self.language_name}测试集真实类别分布', fontsize=13, fontweight='bold')
        ax1.grid(True, alpha=0.3, axis='y')
        
        # 添加数量标签
        for bar, count in zip(bars1, true_counts):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + 10, 
                    f'{count}', ha='center', va='bottom', fontweight='bold')
        
        # 右图：预测分布
        pred_labels = ['预测为Negative', '预测为Positive']
        pred_counts = [pred_negative, pred_positive]
        pred_colors = ['lightcoral', 'lightsalmon']
        
        bars2 = ax2.bar(pred_labels, pred_counts, color=pred_colors, edgecolor='black', alpha=0.8)
        ax2.set_xlabel('预测类别', fontsize=12)
        ax2.set_ylabel('样本数量', fontsize=12)
        ax2.set_title(f'{self.language_name}模型预测类别分布', fontsize=13, fontweight='bold')
        ax2.grid(True, alpha=0.3, axis='y')
        
        # 添加数量标签
        for bar, count in zip(bars2, pred_counts):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 10, 
                    f'{count}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
    
    def generate_performance_report(self, save_dir: str = './performance_report'):
        """生成完整的性能评估报告"""
        # 创建保存目录
        os.makedirs(save_dir, exist_ok=True)
        
        print("="*70)
        print(f"{self.language_name}情感分析LSTM模型性能评估报告")
        print("="*70)
        
        # 打印基本信息
        print(f"\n1. 模型基本信息:")
        print(f"   - 语言: {self.language_name}")
        print(f"   - 测试集大小: {len(self.y_true)} 条评论")
        print(f"   - 测试损失: {self.test_loss:.4f}")
        
        # 打印混淆矩阵
        print(f"\n2. 混淆矩阵:")
        print(f"   [[{self.cm[0,0]:4d} {self.cm[0,1]:4d}]  # Negative类别")
        print(f"    [{self.cm[1,0]:4d} {self.cm[1,1]:4d}]]  # Positive类别")
        
        # 打印详细解释
        print(f"\n   混淆矩阵解释:")
        print(f"   - TN(真负): {self.metrics['true_negative']} - 正确预测为Negative的Negative样本")
        print(f"   - FP(假正): {self.metrics['false_positive']} - 错误预测为Positive的Negative样本")
        print(f"   - FN(假负): {self.metrics['false_negative']} - 错误预测为Negative的Positive样本")
        print(f"   - TP(真正): {self.metrics['true_positive']} - 正确预测为Positive的Positive样本")
        
        # 打印性能指标
        print(f"\n3. 性能指标:")
        print(f"   - 准确率 (Accuracy): {self.metrics['accuracy']:.4f} ({self.metrics['accuracy']*100:.2f}%)")
        print(f"   - 误分类率: {self.metrics['error_rate']:.4f}")
        
        print(f"\n4. 各类别详细指标:")
        print(f"   Negative类别:")
        print(f"     - 精确率: {self.metrics['precision_negative']:.4f}")
        print(f"     - 召回率: {self.metrics['recall_negative']:.4f}")
        print(f"     - F1分数: {self.metrics['f1_negative']:.4f}")
        
        print(f"\n   Positive类别:")
        print(f"     - 精确率: {self.metrics['precision_positive']:.4f}")
        print(f"     - 召回率: {self.metrics['recall_positive']:.4f}")
        print(f"     - F1分数: {self.metrics['f1_positive']:.4f}")
        
        print(f"\n5. 宏平均指标:")
        print(f"   - 宏平均精确率: {self.metrics['precision_macro']:.4f}")
        print(f"   - 宏平均召回率: {self.metrics['recall_macro']:.4f}")
        print(f"   - 宏平均F1分数: {self.metrics['f1_macro']:.4f}")
        
        # 打印分类报告
        print(f"\n6. 详细分类报告:")
        report = self.generate_classification_report()
        print(report)
        
        # 打印误分类分析
        print(f"\n7. 误分类分析:")
        print(f"   - 总体误分类率: {self.metrics['error_rate']:.4f} ({self.metrics['error_rate']*100:.2f}%)")
        print(f"   - 将Negative误判为Positive的比例: {self.metrics['false_positive_rate']:.4f} ({self.metrics['false_positive_rate']*100:.2f}%)")
        print(f"   - 将Positive误判为Negative的比例: {self.metrics['false_negative_rate']:.4f} ({self.metrics['false_negative_rate']*100:.2f}%)")
        
        # 模型性能评估
        print(f"\n8. 模型性能评估:")
        accuracy = self.metrics['accuracy']
        
        if accuracy >= 0.85:
            rating = "优秀"
            suggestion = "模型性能非常好，可以直接部署使用"
        elif accuracy >= 0.75:
            rating = "良好"
            suggestion = "模型性能良好，可以考虑进一步优化以提升性能"
        elif accuracy >= 0.65:
            rating = "一般"
            suggestion = "模型性能一般，建议进行调优或增加训练数据"
        elif accuracy >= 0.55:
            rating = "较差"
            suggestion = "模型性能较差，需要重新设计模型或增加大量训练数据"
        else:
            rating = "差"
            suggestion = "模型性能不理想，建议重新评估数据质量和模型架构"
        
        print(f"   - 准确率等级: {rating} ({accuracy*100:.2f}%)")
        print(f"   - 建议: {suggestion}")
        
        # 类别平衡性评估
        precision_diff = abs(self.metrics['precision_negative'] - self.metrics['precision_positive'])
        recall_diff = abs(self.metrics['recall_negative'] - self.metrics['recall_positive'])
        
        if max(precision_diff, recall_diff) < 0.1:
            balance = "平衡"
        elif max(precision_diff, recall_diff) < 0.2:
            balance = "基本平衡"
        else:
            balance = "不平衡"
        
        print(f"   - 类别平衡性: {balance}")
        
        # 保存结果
        print(f"\n9. 生成可视化图表...")
        
        # 创建结果字典
        results = {
            'language': self.language,
            'language_name': self.language_name,
            'test_loss': float(self.test_loss),
            'confusion_matrix': self.cm.tolist(),
            'metrics': self.metrics,
            'classification_report': report,
            'performance_evaluation': {
                'accuracy_rating': rating,
                'suggestion': suggestion,
                'class_balance': balance
            }
        }
        
        # 转换为可序列化格式
        results_serializable = convert_to_serializable(results)
        
        # 保存为JSON
        json_path = os.path.join(save_dir, f'{self.language}_performance_report.json')
        with open(json_path, 'w', encoding='utf-8') as f:
            json.dump(results_serializable, f, ensure_ascii=False, indent=2)
        print(f"   详细报告已保存到: {json_path}")
        
        # 生成可视化图表
        self.plot_confusion_matrix(os.path.join(save_dir, f'{self.language}_confusion_matrix.png'))
        self.plot_performance_metrics(os.path.join(save_dir, f'{self.language}_performance_metrics.png'))
        self.plot_error_analysis(os.path.join(save_dir, f'{self.language}_error_analysis.png'))
        self.plot_class_distribution(os.path.join(save_dir, f'{self.language}_class_distribution.png'))
        
        # 创建总结CSV
        summary_data = []
        for key, value in self.metrics.items():
            summary_data.append({
                '指标': key,
                '值': f'{value:.6f}' if isinstance(value, float) else str(value)
            })
        
        df = pd.DataFrame(summary_data)
        csv_path = os.path.join(save_dir, f'{self.language}_performance_summary.csv')
        df.to_csv(csv_path, index=False, encoding='utf-8-sig')
        print(f"   总结CSV已保存到: {csv_path}")
        
        print(f"\n" + "="*70)
        print("性能评估完成！")
        print(f"所有结果已保存到: {save_dir}")
        print("="*70)
        
        return results_serializable

def main():
    """主函数：评估模型性能"""
    print("="*70)
    print("LSTM情感分析模型性能评估")
    print("基于提供的识别结果进行分析")
    print("="*70)
    
    # 已提供的识别结果
    test_loss = 0.5701
    confusion_matrix_data = [[905, 345], [274, 976]]
    
    # 评估中文模型
    print("\n1. 评估中文模型性能...")
    print("-" * 50)
    
    cn_evaluator = ModelPerformanceEvaluator(
        test_loss=test_loss,
        confusion_matrix_data=confusion_matrix_data,
        language='cn'
    )
    
    cn_results = cn_evaluator.generate_performance_report(save_dir='./performance_report/cn')
    
    # 评估英文模型（假设有类似结果）
    print("\n\n2. 评估英文模型性能（示例）...")
    print("-" * 50)
    
    # 假设英文模型有类似但略好的表现
    en_confusion_matrix_data = [[920, 330], [260, 990]]  # 假设值
    en_test_loss = 0.5501  # 假设值
    
    en_evaluator = ModelPerformanceEvaluator(
        test_loss=en_test_loss,
        confusion_matrix_data=en_confusion_matrix_data,
        language='en'
    )
    
    en_results = en_evaluator.generate_performance_report(save_dir='./performance_report/en')
    
    # 模型对比
    print("\n\n3. 中英文模型对比分析")
    print("="*50)
    
    comparison_data = {
        '模型': ['中文模型', '英文模型'],
        '测试损失': [cn_results['test_loss'], en_results['test_loss']],
        '准确率': [cn_results['metrics']['accuracy'], en_results['metrics']['accuracy']],
        '精确率(Negative)': [cn_results['metrics']['precision_negative'], en_results['metrics']['precision_negative']],
        '召回率(Negative)': [cn_results['metrics']['recall_negative'], en_results['metrics']['recall_negative']],
        '精确率(Positive)': [cn_results['metrics']['precision_positive'], en_results['metrics']['precision_positive']],
        '召回率(Positive)': [cn_results['metrics']['recall_positive'], en_results['metrics']['recall_positive']],
        '宏平均F1分数': [cn_results['metrics']['f1_macro'], en_results['metrics']['f1_macro']]
    }
    
    df_comparison = pd.DataFrame(comparison_data)
    print("\n性能指标对比:")
    print(df_comparison.to_string(index=False, float_format=lambda x: f'{x:.4f}'))
    
    # 保存对比结果
    comparison_dir = './performance_report/comparison'
    os.makedirs(comparison_dir, exist_ok=True)
    
    df_comparison.to_csv(os.path.join(comparison_dir, 'model_comparison.csv'), 
                         index=False, encoding='utf-8-sig')
    
    # 绘制对比图
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(2)
    width = 0.15
    
    metrics_to_compare = ['测试损失', '准确率', '精确率(Negative)', '召回率(Negative)', 
                         '精确率(Positive)', '召回率(Positive)']
    metric_indices = [1, 2, 3, 4, 5, 6]  # DataFrame中的列索引
    
    for i, (metric, idx) in enumerate(zip(metrics_to_compare, metric_indices)):
        values = comparison_data[list(comparison_data.keys())[idx]]
        offset = (i - 2.5) * width
        bars = ax.bar(x + offset, values, width, label=metric)
        
        # 添加数值标签
        for bar, value in zip(bars, values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                   f'{value:.4f}', ha='center', va='bottom', fontsize=8)
    
    ax.set_xlabel('模型', fontsize=12)
    ax.set_ylabel('分数', fontsize=12)
    ax.set_title('中英文模型性能对比', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(comparison_data['模型'])
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig(os.path.join(comparison_dir, 'model_comparison.png'), 
                dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    
    print(f"\n对比结果已保存到: {comparison_dir}")
    
    # 最终总结
    print("\n" + "="*70)
    print("评估总结")
    print("="*70)
    
    print(f"\n中文模型:")
    print(f"  - 准确率: {cn_results['metrics']['accuracy']*100:.2f}%")
    print(f"  - 性能等级: {cn_results['performance_evaluation']['accuracy_rating']}")
    print(f"  - 类别平衡性: {cn_results['performance_evaluation']['class_balance']}")
    
    print(f"\n英文模型:")
    print(f"  - 准确率: {en_results['metrics']['accuracy']*100:.2f}%")
    print(f"  - 性能等级: {en_results['performance_evaluation']['accuracy_rating']}")
    print(f"  - 类别平衡性: {en_results['performance_evaluation']['class_balance']}")
    
    print(f"\n总体评估:")
    if cn_results['metrics']['accuracy'] > 0.75 and en_results['metrics']['accuracy'] > 0.75:
        print("  ✓ 两个模型都达到了可接受性能水平")
    elif cn_results['metrics']['accuracy'] > 0.75 or en_results['metrics']['accuracy'] > 0.75:
        print("  ⚠ 一个模型性能良好，另一个需要改进")
    else:
        print("  ✗ 两个模型都需要进一步优化")
    
    print(f"\n建议:")
    print(f"  1. {cn_results['performance_evaluation']['suggestion']}")
    print(f"  2. {en_results['performance_evaluation']['suggestion']}")
    
    print(f"\n" + "="*70)
    print("所有评估结果已生成完毕！")
    print(f"详细报告位于: ./performance_report/")
    print("="*70)

if __name__ == "__main__":
    # 运行评估
    main()
```
